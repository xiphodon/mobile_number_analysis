```python
def Calresult(predLabel, testLabel):
    TP = 0
    FP = 0
    TN = 0
    FN = 0
    for i in range(len(predLabel)):
        if testLabel[i] == 0:
            if predLabel[i] == 0:
                TN += 1
            else:
                FP += 1
        else:
            if predLabel[i] == 1:
                TP += 1
            else:
                FN += 1
                
    if (TP + FN) == 0:
        TPR = 0
    else:
        TPR = TP / float(TP + FN)
        
    if (FP + TN) == 0:
        FPR = 0
    else:
        FPR = FP / float(FP + TN)
        
    return (TPR, FPR)
```

```python
def ks_main(possibility, testLabel):
    
    length = len(possibility)
    # Sorting the predict result label and orgional label according to predicted possibility
    sort_dict = {}
    
    for i in range(length):
        sort_dict[i] = (possibility[i], testLabel[i])
        
    sorted_dict = sorted(sort_dict.items(), key = lambda item: item[1][0], reverse = True)


    sorted_possibility = [0] * length
    sorted_testLabel = [0] * length
    
    # Get re-ordered Perdict label and test label 
    for i in range(length):
        sorted_possibility[i] = sorted_dict[i][1][0]
        sorted_testLabel[i]  = sorted_dict[i][1][1]

    TPR_list = [0] * length
    FPR_list = [0] * length
    cal_list = [0] * length
    
    max_ks = 0
    cut_off = 0    
    for i in range(length):
        
        cal_list[0:i+1] = [1 for i in range(i+1)]
        
        TPR_list[i] = Calresult(cal_list, sorted_testLabel)[0]
        FPR_list[i] = Calresult(cal_list, sorted_testLabel)[1]
        
        
        if (TPR_list[i] - FPR_list[i]) >= max_ks:
            max_ks = TPR_list[i] - FPR_list[i]
            cut_off = sorted_possibility[i]
            mx_index = i
    
    cut_off =np.log((1 - cut_off)/cut_off) * 50 / np.log(2) + 500
    return (TPR_list, FPR_list, max_ks, cut_off, mx_index)
```

```python
def Score(defaultprobility):
    return map(lambda x:np.log((1 - x)/x) * 50 / np.log(2) + 500,defaultprobility)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, model_selection, ensemble

X_train, X_test, y_train, y_test = model_selection.train_test_split(df4, df4_target,test_size=0.25, random_state=0,stratify=df4_target)

print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
%matplotlib inline

clf = ensemble.RandomForestClassifier(n_estimators=200,max_depth=10,min_samples_leaf=5,random_state=0)
clf.fit(X_train, y_train)
```

```python
predict = clf.predict_proba(X_train)
a = predict[:, 1]
c = y_train.values
score=Score(a)
[TPR_line, FPR_line, ks, cut_off,mx_index] = ks_main(a,c)

score_sort = sorted(score)

plt.plot(score_sort,TPR_line,'b',score_sort,FPR_line,'g')
plt.plot([cut_off,cut_off],[TPR_line[mx_index],FPR_line[mx_index]],'r-')
plt.show()
print('max K-S = %f' %ks)
print('cut-off value = %.2f' %cut_off)
```

```python
def my_score(estimator,x,y):
    a = estimator.predict_proba(x)[:,1]
    c = y.values
#     score=Score(a)
    [TPR_line, FPR_line, ks, cut_off,mx_index] = ks_main(a,c)
    return ks
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import cross_val_score
# 使用xgboost自带方法模型
n_estimators_list = range(50,60,20)
max_depth_list = range(3,4)
min_samples_leaf_list = range(9,10,2)
max_features_list = [i/10.0 for i in range(9,11)]
test_scores = []
params_eta_index = []
for n_estimators in n_estimators_list:
    for max_depth in max_depth_list:
        for min_samples_leaf in min_samples_leaf_list:
            for max_features in max_features_list:
                clf = ensemble.RandomForestClassifier(n_estimators=n_estimators,max_depth=max_depth,min_samples_leaf=min_samples_leaf,max_features=max_features)
                ks_list = cross_val_score(clf, df4, df4_target, cv=5, scoring=my_score)
                test_scores.append(np.mean(ks_list))
                params_eta_index.append((n_estimators,max_depth,min_samples_leaf,max_features))
                print(n_estimators,max_depth,min_samples_leaf,max_features,np.mean(ks_list))
```
